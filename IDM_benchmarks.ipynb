```python
vev = 246
l1 = 0.129

def plus(m,dplus):
    plus = dplus*(2.*m + dplus)/(vev**2.)
    return plus

def zero(m,dzero):
    zero = dzero*(2.*m + dzero)/(vev**2.)
    return zero

##############

def c1(m,dplus,dzero):
    c1 = (8.*np.pi - zero(m,dzero) )/2.
    return np.array(c1)

def c2(m,dplus,dzero):
    c2 = (8.*np.pi + zero(m,dzero) - 4.*plus(m,dplus) )/2.
    return np.array(c2)

def c3(m,dplus,dzero):
    c3 = (8.*np.pi*np.sqrt(1-0.75*l1/np.pi) - zero(m,dzero) - 2.*plus(m,dplus)  )/4.
    return np.array(c3)

def c6(m,dplus,dzero):
    c6 = (8.*np.pi - zero(m,dzero) - 2.*plus(m,dplus)  )/2.
    return np.array(c6)


def c10(m,dplus,dzero):
    c10 = (8.*np.pi - 5.*zero(m,dzero) + 2.*plus(m,dplus)   )/2.
    return np.array(c10)





```

## Unitarity mass upper limit 

```python
vev = 246
l1 = 0.129

def plus(m,dplus):
    plus = dplus*(2.*m + dplus)/(vev**2.)
    return plus

def zero(m,dzero):
    zero = dzero*(2.*m + dzero)/(vev**2.)
    return zero

##############

def c1c(l,m,dplus,dzero):
    c1 = - 2.*l + 8.*np.pi - zero(m,dzero) 
    return np.array(c1)

def c2c(l,m,dplus,dzero):
    c2 = 8.*np.pi + zero(m,dzero) - 4.*plus(m,dplus) -2.*l
    return np.array(c2)

def c3c(l,m,dplus,dzero):
    c3 = 8.*np.pi*np.sqrt(1-0.75*l1/np.pi) - zero(m,dzero) - 2.*plus(m,dplus)  - 4.*l
    return np.array(c3)

def c6c(l,m,dplus,dzero):
    c6 = 8.*np.pi - zero(m,dzero) - 2.*plus(m,dplus) - 2.*l
    return np.array(c6)


def c10c(l,m,dplus,dzero):
    c10 = 8.*np.pi - 5.*zero(m,dzero) + 2.*plus(m,dplus)   - 2.*l
    return np.array(c10)
```
